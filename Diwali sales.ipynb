```python
State
```

```python
df.columns
```

```python
# Total number of order from top 10 statess
sale_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
#set  the size of bar
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data = sale_state, x= 'State', y= 'Orders')
```

```python
# Total amount / sales from top 10 states
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)

sns.set(rc = { 'figure.figsize': (15,5)})
sns.barplot(data = sales_state , x = 'State' , y = 'Amount')
```

```python
From above graphs we can see that most of the orders & total sales / amount are from Uttar Pradesh, Maharashtra and Karnataka 
```
